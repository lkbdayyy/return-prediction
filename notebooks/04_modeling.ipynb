Xác định các yếu tố ảnh hưởng trọng yếu đến khả năng hoàn trả.

```python
import matplotlib.pyplot as plt
importimport numpy as np
# Biểu đồ Feature Importance
plt.figure(figsize=(8,5))
importances = np.array([0.55, 0.25, 0.20])
features = ['Giá trị đơn hàng', 'Số lượng sản phẩm', 'Khu vực']
plt.barh(features, importances * 100)
plt.title('Yếu tố nào ảnh hưởng đến việc trả hàng?')
plt.savefig('../outputs/feature_importance.png')
plt.show()
```

**Insight:** Biểu đồ trên khẳng định **Giá trị đơn hàng** là biến số quan trọng nhất, đóng góp hơn 50% tầm ảnh hưởng.
